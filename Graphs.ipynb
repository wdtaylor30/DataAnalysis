```python
# import necessary libraries
import seaborn as sns
from matplotlib import pyplot as plt
import pandas as pd
import math

%matplotlib inline
```

```python
# import dataset
tideData = pd.read_excel("/home/wdtay/Programming/Projects/Tides/Data/Tide_Data_Analysis.xlsx")
```

```python
# assign dates to list
date = []
for i in range(1, 32):
    date.append(i)
```

```python
# assign Low Tide 1's heights to variable
lowTide = tideData['Height (ft)']

# assign High Tide 1's heights to variable
highTide = tideData['Height (ft).2']
```

```python
# create scatterplots sans regression model

# low tide
plt.xlabel("Day of Month")
plt.ylabel("Tide Height")
plt.title("Low Tide Height\nCape Hatteras, North Carolina\nMarch of 2017")
sns.set_style('darkgrid')
plt.plot(date, lowTide, 'bo')
plt.show()

# high tide
plt.xlabel("Day of Month")
plt.ylabel("Tide Height")
plt.title("High Tide Height\nCape Hatteras, North Carolina\nMarch of 2017")
sns.set_style('darkgrid')
plt.plot(date, highTide, 'bo')
plt.show()
```

```python
# input regression model and graph wave
def lowTideReg(x):
    height = .3939075012 * math.sin(.4129361724 * x - .4196480264) - .1147194454
    return height

# list to hold tide heights
lowTideRegHeights = []

# calculate height for each day and populate list
for i in range(0, 31):
    lowTideRegHeights.append(lowTideReg(i))
    
# populate new dataframe (dataset) with predicted heights
lowTideRegression = pd.DataFrame({'Height': lowTideRegHeights}, index = date)

# graph
sns.set_style('darkgrid')
plt.xlabel("Day of Month")
plt.ylabel("Tide Height (ft)")
plt.title("Low Tide Height with Sine Regression\nCape Hatteras, North Carolina\nMarch 2017")
plt.plot(date, lowTide, 'bo', label = 'Actual Data')
plt.plot(date, lowTideRegression, 'g', label = 'Regression Model')
plt.legend()
```

```python
# repeat process for high tide
def highTideReg(x):
    height = .6210783029 * math.sin(.3824004589 * x - 2.922834346) + 2.926210734
    return height

# list to hold tide heights
highTideRegHeights = []

# calculate height for each day and populate list
for i in range(0, 31):
    highTideRegHeights.append(highTideReg(i))
    
# populate new dataframe (dataset) with predicted heights
highTideRegression = pd.DataFrame({'Height': highTideRegHeights}, index = date)

# graph
sns.set_style('darkgrid')
plt.xlabel("Day of Month", )   
plt.ylabel("Tide Height (ft)") 
plt.title("High Tide Height with Sine Regression\nCape Hatteras, North Carolina\nMarch 2017")
plt.plot(date, highTide, 'bo', label = 'Actual Data')
plt.plot(date, highTideRegression, 'g', label = 'Regression Model')
plt.legend()
```
